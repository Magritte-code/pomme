# Function accuracy

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

## Cancelation error in $\exp\left(-\left(a+\delta\right)^{2}\right) - \exp\left(-a^{2}\right)$

```python
delta = 1.0e-4
width = 1.0e+1
a = torch.linspace(-width, +width, 1000)
# function
b = torch.erf(a+delta) - torch.erf(a)
# series expansion
s = 2.0 * torch.exp(-a**2) / np.sqrt(np.pi) * delta * (1 - a*delta + 1/3 * (-1+2*a**2)*delta**2)

fig, axs = plt.subplots(2,1, dpi=150)
axs[0].plot(a, b)
axs[0].plot(a, s)
axs[0].set_yscale('log')
axs[1].plot(a, 2.0*torch.abs(b-s)/(b+s))
axs[1].set_yscale('log')
```

## Cancelation error in $\text{Erf}\left(a+\delta\right) - \text{Erf}\left(a\right)$

```python
delta = 1.0e-5
width = 1.0e+1
a = torch.linspace(-width, +width, 1000)
# function
b = torch.exp(-(a+delta)**2) - torch.exp(-a**2)
# series expansion
s = torch.exp(-a**2) * (-2*a*delta + (-1+2*a**2)*delta**2)

fig, axs = plt.subplots(2,1, dpi=150)
axs[0].plot(a, b)
axs[0].plot(a, s)
# axs[0].set_yscale('log')
axs[1].plot(a, 2.0*torch.abs(b-s)/(b+s))
axs[1].set_yscale('log')
```

## Cancelation error in $\exp\left(a+\delta\right)-\left(1-\delta\right)\exp\left(a\right)$

```python
delta = 1.0e-4
width = 1.0e+0
a = torch.linspace(-width, +width, 1000, dtype=torch.float64)
# function
b = torch.exp(a+delta) - (1 + delta) * torch.exp(a)
# series expansion
s = torch.exp(a) * delta**2 * (1/2 + 1/6*delta + 1/120*delta**2)

fig, axs = plt.subplots(2,1, dpi=150)
axs[0].plot(a, b)
axs[0].plot(a, s)
axs[0].set_yscale('log')
axs[1].plot(a, 2.0*torch.abs(b-s)/(b+s))
axs[1].set_yscale('log')
```
