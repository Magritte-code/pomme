# Synthetic observations - spherical model
---

```python
import matplotlib.pyplot as plt
import torch

from spherical import lines, velocities, frequencies, smodel, r_out
```

```python
obss = smodel.image(lines, frequencies, r_max=r_out)
```

```python
plt.figure(dpi=150)
for obs in obss:
    plt.plot(velocities, obs.data)
```
